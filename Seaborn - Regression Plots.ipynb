**REGRESSION PLOT**

```python
import seaborn as sns
tips_data = sns.load_dataset('tips')
tips_data.head(5)
```

```python
#Direct plot
sns.lmplot(x='total_bill', y ='tip', data = tips_data)
```

```python
#With Hue
sns.lmplot(x='total_bill', y ='tip', data = tips_data,
          hue = 'sex', markers = ['o','*'])
```

```python
#Defining size of data points
sns.lmplot(x='total_bill', y ='tip', data = tips_data,
          hue = 'sex', markers = ['o','*'],
          scatter_kws = {'s':100, 'linewidth':0.5})
```

```python
#Creating Grid Plots
sns.lmplot(x='total_bill', y ='tip', data = tips_data, row ='smoker', col = 'time')
```
