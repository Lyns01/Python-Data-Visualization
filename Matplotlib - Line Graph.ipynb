```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Simple Plot
x = [1,2,3]
y =[2,4,6]
plt.plot(x,y) 
#plt.show() #To get rid of that line - [<matplotlib.lines.Line2D at 0x2693df58070>] - use show at the end
```

```python
#Titles, X and Y label

plt.plot(x,y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('The Graph', fontdict = {'fontsize' : 20}) #Under Font dict you can mention font, color allignment and more - check documenttaion

plt.show() 
```

```python
#To change the X Ticks and Y Ticks - Important - Can change the way how our graph looks
plt.plot(x,y)

plt.xticks([1,2,3])
plt.yticks([1,2,3,4,5,6]) 

plt.show() 
```

```python
#To add legends - what does the line represent - equation or label in case many lines are there
plt.plot(x,y, label = '2x')

plt.xticks([1,2,3])
plt.yticks([1,2,3,4,5,6]) 

plt.legend()

plt.show() 
```

```python
#Format the dataline more informative and look good
plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, linestyle ='--', marker = '*', markersize = 10, markeredgecolor ='blue')
#All these params can be found in documentation 
#plt.plot(x,y,'b*-',label='2x') #[color,marker,line] #The short hand notation works only after the x and y parameter

plt.xticks([1,2,3])
plt.yticks([1,2,3,4,5,6]) 

plt.legend()

plt.show() 
```

```python
#Add Another Line 
x1 = np.arange(1,10,0.5)
y1 = [a**2 for a in x1]
x2 = np.arange(2,9,0.4)
y2 = [a*4+15 for a in x1]
```

```python
#You can split the same line into two different graphs and plot

#Set a plot size
plt.figure(figsize=(3,3),dpi=100) #1200*1200 pixels -> try to keep pixels as high as possible, when you use it as an image - it would be blurry if its less

plt.plot(x1[:13],y1[:13],label = 'x1 y1', marker='*', markersize=10)
plt.plot(x1[12:],y1[12:],label = 'x1 y1', linestyle = '--', marker='*', markersize=10)
plt.plot(x2,y2,label = 'x2 y2')

plt.title('Two Graphs')
plt.xlabel('X1,X2')
plt.ylabel('Y1,Y2')
plt.legend()

#Save it as an image 
plt.savefig('mygraph.png',dpi=300) #U can mention the dpi parameter here to save it at high res

plt.show()
```

```python
#Real world data sample

#Plot a graph

gas = pd.read_csv('gas_prices.csv')
gas.head(5)

x = gas['Year']
y1 = gas['UK']
y2 = gas['USA']
y3 = gas['Canada']

plt.figure(figsize=(5,3),dpi=100)

plt.plot(x,y1,'*-',label ='UK',) 
plt.plot(x,y2,label ='USA')
plt.plot(x,y3,label ='Canada')

plt.title('Gas Prices in UK and USA')
plt.ylabel('Price')
plt.legend()

plt.show()
```

```python
#To do it for a lot of countries

plt.figure(figsize=(7,7),dpi=100)

countries = list(gas.columns)
countries.remove('Year')

#plt.plot(gas['Year'],gas[countries[2]])
for country in countries:
    plt.plot(gas['Year'],gas[country], label = country, marker ='*')
    
plt.title('Gas Prices in UK and USA')
plt.ylabel('Price')
plt.legend()

plt.show()
```
