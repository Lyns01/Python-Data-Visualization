```python
import seaborn as sns
```

```python
tip_data = sns.load_dataset('tips')
tip_data.head(5)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tip_data)
```

```python
sns.countplot(x = 'sex',data = tip_data)
```

```python
sns.boxplot(x = 'day', y = 'tip', data = tip_data, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tip_data, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tip_data, hue = 'sex', split = 'true')
```

```python
sns.stripplot(x = 'day', y = 'tip', data = tip_data, hue = 'sex', dodge = 'true')
```

```python
sns.violinplot(x = 'day', y = 'tip', data = tip_data, hue = 'sex', split = 'true')
sns.swarmplot(data= tip_data, x = 'day', y ='tip', color = 'white')
```
