```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa_data = pd.read_csv('fifa_data.csv')
fifa_data.head(5)
```

```python
#Default Box Plot
Barcelona = fifa_data[fifa_data['Club'] == 'FC Barcelona']['Overall']
Madrid = fifa_data[fifa_data['Club'] == 'Real Madrid']['Overall']

values = [Barcelona,Madrid]
labels = ['Barcelona', 'Madrid']

plt.boxplot(values, labels = labels)

plt.title("Comparison")
plt.xlabel('Clubs')
plt.ylabel('Score')

plt.show()
```

```python
#Colored Box Plot
Barcelona = fifa_data[fifa_data['Club'] == 'FC Barcelona']['Overall']
Madrid = fifa_data[fifa_data['Club'] == 'Real Madrid']['Overall']

values = [Barcelona,Madrid]
labels = ['Barcelona', 'Madrid']

boxes = plt.boxplot(values, 
                    labels = labels,
                   patch_artist = True, #Fill color in
                   medianprops = {'linewidth' : 2}) 

plt.title("Comparison")
plt.xlabel('Clubs')
plt.ylabel('Score')

plt.show()
```

```python
#To change default fill in / edge colors
Barcelona = fifa_data[fifa_data['Club'] == 'FC Barcelona']['Overall']
Madrid = fifa_data[fifa_data['Club'] == 'Real Madrid']['Overall']

values = [Barcelona,Madrid]
labels = ['Barcelona', 'Madrid']

boxes = plt.boxplot(values, 
                    labels = labels,
                   patch_artist = True, #Fill color in
                   medianprops = {'linewidth' : 2}) 

for box in boxes['boxes'] :
    #Change default edge color
    box.set(color = '#0000FF')
    #Change Fill color
    box.set(facecolor='#00FF00')
    
plt.title("Comparison")
plt.xlabel('Clubs')
plt.ylabel('Score')

plt.show()
```
