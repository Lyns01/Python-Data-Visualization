```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**HEATMAPS**

```python
print(sns.get_dataset_names())
```

```python
#Read the dataset
crash_dataset = sns.load_dataset('car_crashes')
crash_dataset.head(5)
```

```python
#Get the covariance matrix - since this has a lot of rcolumns, find which ones are highy corelated
crash_mat = crash_dataset.corr(numeric_only=True) 
#if you dont add this parameter it will throw error as it does not eliminate cat columns by default
```

```python
#Create a heatmat
sns.heatmap(crash_mat)
```

```python
sns.heatmap(crash_mat, annot = True) #Adding values to it
```

```python
sns.heatmap(crash_mat, annot = True, cmap = "Blues") #cmap / color maps in matplotlib documentation
```

```python
#Read Data
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
#Create a pivot table
flights_mat = flights.pivot_table(index = 'month', columns ='year', values = 'passengers')
flights_mat
```

```python
#Create a heatmap
plt.figure(figsize = (5,6))
sns.heatmap(flights_mat,cmap='Blues',linecolor = 'red',linewidth = 1)
```

**CONFUSION MATRIX** - A form of heatmap

```python
cm = np.array([[1,10],[2,9]])
cm
```

```python
sns.heatmap(cm,annot = True,cmap = 'Greens') 
```

```python
#Bar on the right is called color bar - To Modify that
sns.heatmap(cm,annot = True,cmap = 'Greens',cbar_kws = {'orientation':'horizontal', 'label':'color bar'}) 
```

```python
#Change the labels - this doesnt change the graph - just the labels
sns.heatmap(cm,annot = True,cmap = 'Greens',xticklabels=[10,11],yticklabels=[20,30]) 
plt.xlabel('actual')
plt.ylabel('predicted')
```

**PAIR GRID**

```python
#Simillar to pair plot but this gives more control over 
# 1. The type of the graph 
# 2. The data columns to be considered
iris_data = sns.load_dataset('iris')
iris_data.head(5)
```

```python
sns.pairplot(iris_data,hue='species') #Cant change the type of graph of data columns
```

```python
#Scatter Plot
iris_plot = sns.PairGrid(iris_data,hue='species')
iris_plot.map(plt.scatter)
```

```python
#You can change diag and off diag plots
iris_plot = sns.PairGrid(iris_data,hue='species')
iris_plot.map_diag(plt.hist)
iris_plot.map_offdiag(plt.scatter)
```

```python
#You can change upper lower plots wrt diagonal 
iris_plot = sns.PairGrid(iris_data,hue='species')
iris_plot.map_diag(plt.hist)
iris_plot.map_upper(sns.kdeplot)
iris_plot.map_lower(plt.scatter)
```

```python
#Contorl the data columns
iris_plot = sns.PairGrid(iris_data,hue='species',
                        x_vars = ['sepal_length','sepal_width'],
                         y_vars = ['petal_length', 'petal_width'])
iris_plot.map(plt.scatter)
```

**FACET GRID**

```python
#Multiple plots in a grid
tips_data = sns.load_dataset('tips')
tips_data.head(5)
```

```python
#So we have the time of the day (Lunch and Dinner) and whether they are smoker or not
#To plot the tip against these two values
tip_plot = sns.FacetGrid(tips_data, row ='time', col = 'smoker')
tip_plot.map(plt.hist, 'total_bill', bins = 8) #Hist needs only one arguments
```

```python
#Same with scatter
tip_plot = sns.FacetGrid(tips_data, row ='time', col = 'smoker')
tip_plot.map(plt.scatter, 'total_bill', 'tip') #Scatter needs two arguments
```

```python
#Same with scatter more variations - can also be done with just row/column
tip_plot = sns.FacetGrid(tips_data, col ='time',col_order = ['Dinner','Lunch'], hue = 'smoker') 
tip_plot.map(plt.scatter, 'total_bill', 'tip') #Scatter needs two arguments
```

```python
#To pass more arguments to the map function and to mention the rep symbols for data points
tips_plot = sns.FacetGrid(tips_data,
                          col ='sex',
                          hue = 'smoker', hue_order = ['Yes','No'],
                          height = 4,
                          aspect = 1.3,
                          hue_kws = {'marker':['*','^']}
                         )
kws = dict(s=50, linewidth = .5, edgecolor ='w')
tips_plot.map(plt.scatter, 'total_bill', 'tip', **kws)
```
