```python
import seaborn as sns #Seaborn build on top of matplotlib #Easy to call functions when compared to matplotlib
import matplotlib.pyplot as plt

%matplotlib inline #To show all plots inline
%reload_ext autoreload #To reload every 2 secs 
%autoreload 2
```

```python
#There are a few dataset available in seaborn library
print(sns.get_dataset_names())
```

```python
car_crashes = sns.load_dataset('car_crashes')
```

```python
car_crashes.head(5)
```

**DISTRIBUTION PLOT**

```python
#Distribution plot - Univariate distribution - one variable 
#Pressing Shift + Tab with cursor on the finction shows the parameters 
#sns.distplot(car_crashes['not_distracted'])

#The bell shaped curve is the kernal density function you can eliminate it if you want 
#sns.distplot(car_crashes['not_distracted'], kde = False)

#Introduce bins - divide data into differnent intervals 
sns.distplot(car_crashes['not_distracted'], kde = False, bins = 25)
```

**JOINT PLOT**

```python
#
#All plot function wants x y and data
#sns.jointplot(x='speeding',y = 'alcohol', data = car_crashes)
#To Include the kernal density function, add the kind of graph required - kde, reg, hex
sns.jointplot(x='speeding',y = 'alcohol', data = car_crashes, kind ='reg')
```

**KDE PLOT**

```python
sns.kdeplot(car_crashes['speeding'])
```

**PAIR PLOT**

```python
#Compares all the data columns with each other
tips_data = sns.load_dataset('tips')
tips_data.head(5)
```

```python
sns.pairplot(data = tips_data,hue = 'sex')
```

**RUG PLOT**

```python
#Just like a historgram - there its taller, here its thicker 
#1D Scatter plot - shows where the data point is more and less - or the distribution of the data points
sns.rugplot(tips_data['tip'])
```

**STYLING**

```python
#BG and Grid
sns.set_style('white') #dark, white, darkgrid, whitegrid

#Figure Size
plt.figure(figsize=(8,4))

#Change the labels how its presented 
sns.set_context('paper', font_scale = 1.4) #paper, poster, talk

#The Plot
sns.jointplot(x='speeding',y = 'alcohol', data = car_crashes, kind ='reg')

#The axis lines
sns.despine(left = True) #The left axis is removed
```

```python
#Box plot will be covered in other notebook - this is just sample for styling option
sns.boxplot(x='day', y ='tip', data = tips_data, hue = 'sex', palette = 'afmhot') #values for palette can be taken from matplotlib documentation
plt.legend(loc = 2) #Position the legends - 1,2,3,4 (4 corners) - can be found in documentation
```
