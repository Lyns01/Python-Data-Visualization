**CATEGORICAL PLOTS**

```python
import seaborn as sns
import numpy as np
tip_data = sns.load_dataset('tips')
tip_data.head(5)
```

**BAR PLOTS**

```python
sns.barplot(x = 'sex',y = 'tip', data = tip_data, estimator = np.mean) 
#by default estimator is mean, but it can be median, std, covar, var
```

**COUNT PLOT**

```python
sns.countplot(x = 'sex', data = tip_data) #Just counts based on the parameter specified
```

**BOX PLOT**

```python
tip_data.head(5)
```

```python
sns.boxplot(x='day',y='tip',data = tip_data, hue='sex')
```

**VIOLIN PLOT**

```python
#Box plot consideres the data points while the violin plot takes the KDE of the data points
sns.violinplot(x='day',y='tip',data = tip_data, hue='sex') 
```

```python
sns.violinplot(x='day',y='tip',data = tip_data, hue='sex', split = True) #puts the hue categories together
```

**STRIP PLOT**

```python
#Scatter plot with one categorical variable
sns.stripplot(data= tip_data, x = 'day', y ='tip')
```

```python
sns.stripplot(data= tip_data, x = 'day', y ='tip', hue='sex')
```

```python
sns.stripplot(data= tip_data, x = 'day', y ='tip', hue='sex', dodge = True)
```

**SWARM PLOT**

```python
#Scatter plot and Strip plot are showing teh actual datapoints while violin and swarm plots shows the kde of the data points
#More number of points in a horizondal line means it has occured more
sns.violinplot(data= tip_data, x = 'day', y ='tip')
sns.swarmplot(data= tip_data, x = 'day', y ='tip', color = 'white')
```
